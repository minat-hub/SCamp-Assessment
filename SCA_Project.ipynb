```
#The best selling books of the decade.
df2.plot(x = "Year", y = "Price", kind = "scatter", marker = "*");
plt.title('Best Selling Books of the decade') 
plt.xlabel('Year')
plt.ylabel('Prices')
```

```
#Visualizing percentage of fiction to non fiction
from collections import Counter
print(Counter(df2['Genre']))
```

```
title_type = df2.groupby('Genre').agg('count')
print(title_type)
```

```
import matplotlib.ticker as ticker
import matplotlib.cm as cm
import matplotlib as mpl
from matplotlib.gridspec import GridSpec
import matplotlib.pyplot as plt
type_labels = title_type.Name.sort_values().index 
type_counts = title_type.Name.sort_values()
plt.figure(1, figsize=(20,10)) 
the_grid = GridSpec(2, 2)
cmap = plt.get_cmap('Spectral')
colors = [cmap(i) for i in np.linspace(0, 1, 8)]
plt.subplot(the_grid[0, 1], aspect=1, title='Percentage of fiction to non fiction')
type_show_ids = plt.pie(type_counts, labels=type_labels, autopct='%1.1f%%', shadow=True, colors=colors)
plt.show()
```

```
#Trends over the years will include the following 
#Best Selling Book Price Declines
#Fiction Book Price Overtime
#Non Fiction Book Prices Overtime
#Reviews of Nonfiction
#Reviews of Fiction
#Ratings Over Time
#Non Fiction Ratings Over Time
#Fiction Ratings Over Time
```

```
#Best Selling Book Price Declines
df2.groupby('Year').mean().plot(y='Price');
plt.title('Best Selling Book Price Declines') 
plt.xlabel('Year')
plt.ylabel('Prices')
```

```
#Best Selling Book Price Declines
df2[df2.Genre == 'Fiction'].groupby('Year').mean().plot(y='Price');
plt.title('Fiction Book Price Overtime') 
plt.xlabel('Year')
plt.ylabel('Prices')
```

```
#Non Fiction Book Prices Overtime
df2[df2.Genre == 'Non Fiction'].groupby('Year').mean().plot(y='Price');
plt.title('Non Fiction Book Prices Overtime')
plt.xlabel('Year')
plt.ylabel('Prices')
```

```
#Reviews of Nonfiction
df2[df2.Genre == 'Non Fiction'].groupby('Year').mean().plot(y='Reviews');
plt.title('Reviews of Nonfiction') 
plt.xlabel('Year')
plt.ylabel('# of Reviews')
```

```
#Reviews of Fiction
df2[df2.Genre == 'Fiction'].groupby('Year').mean().plot(y='Reviews');
plt.title('Reviews of Fiction') 
plt.xlabel('Year')
plt.ylabel('# of Reviews')
```

```
#Ratings Over Time
df2.groupby('Year').mean().plot(y='User Rating');
plt.title('Ratings Over Time') 
plt.xlabel('Year')
plt.ylabel('User Rating')
```

```
#Non Fiction Ratings Over Time
df2[df2.Genre == 'Non Fiction'].groupby('Year').mean().plot(y='User Rating');
plt.title('Non Fiction Ratings Over Time') 
plt.xlabel('Year')
plt.ylabel('User Rating')
```

```
#Fiction Ratings Over Time
df2[df2.Genre == 'Fiction'].groupby('Year').mean().plot(y='User Rating');
plt.title('Fiction Ratings Over Time') 
plt.xlabel('Year')
plt.ylabel('User Rating')
```

```
#Top 10 authors according to reviews
author_of_most_books = df2.groupby('Author')['Reviews'].count().reset_index().sort_values('Reviews',ascending=False).head(10).set_index('Author')
plt.subplots(figsize=(12,8))
ax = author_of_most_books['Reviews'].sort_values().plot.barh(width=0.9,color=sns.color_palette('gist_rainbow',12))
ax.set_xlabel("Total Reviews ", fontsize=15)
ax.set_ylabel("Authors", fontsize=15)
ax.set_title("Top 10 authors with most reviews",fontsize=20,color='black')
totals = []
for i in ax.patches:
    totals.append(i.get_width())
total = sum(totals)
for i in ax.patches:
    ax.text(i.get_width()+.2, i.get_y()+.2,str(round(i.get_width())), fontsize=15,color='black')
plt.show()
```
